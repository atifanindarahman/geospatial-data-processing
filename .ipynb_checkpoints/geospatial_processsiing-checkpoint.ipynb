```python
import geopandas as gpd
import rasterio
from rasterio.mask import mask
import matplotlib.pyplot as plt
from shapely.geometry import mapping
```

```python
states = gpd.read_file("C:/Users/Atif Aninda/Documents/geospatial-data-processing/data/raw/states.geojson")
states = states[states["name"] == "Colorado"]
```

```python
dem = rasterio.open("C:/Users/Atif Aninda/Documents/geospatial-data-processing/data/raw/dem_co.tif")
states = states.to_crs(dem.crs)
```

```python
geoms = [mapping(geom) for geom in states.geometry]
out_image, out_transform = mask(dem, geoms, crop=True)
```

```python
out_meta = dem.meta.copy()
out_meta.update({
    "height": out_image.shape[1],
    "width": out_image.shape[2],
    "transform": out_transform
})
```

```python
with rasterio.open("C:/Users/Atif Aninda/Documents/geospatial-data-processing/data/processed/dem_clipped.tif", "w", **out_meta) as dest:
    dest.write(out_image)
```

```python
plt.imshow(out_image[0], cmap="terrain")
plt.colorbar(label="Elevation (m)")
plt.savefig("C:/Users/Atif Aninda/Documents/geospatial-data-processing/figures/elevation_map.png", dpi=300)
plt.show()
```
